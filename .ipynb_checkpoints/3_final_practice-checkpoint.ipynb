### 최종 실습 프로젝트

당신은 한 온라인 쇼핑몰의 데이터 분석가로 고용되었다. 쇼핑몰의 판매 데이터를 분석하여 고객의 구매 패턴을 파악하고, 중요한 인사이트를 도출하는 것이 이번 프로젝트의 목표다. 이를 통해 향후 마케팅 전략을 제시하고자 한다. ecommerce_sales_data.csv 파일을 통해 주어진 데이터는 고객의 주문 기록, 상품 정보, 판매 데이터 등을 포함하고 있다. 주어진 데이터를 분석하여 아래의 질문에 답하라.

- **데이터 설명**:
    - **OrderID**: 주문 ID
    - **Product**: 주문한 상품 이름
    - **Category**: 상품의 카테고리 (예: Electronics, Clothing 등)
    - **Quantity**: 주문 수량
    - **Price**: 상품의 가격
    - **CustomerID**: 고객 ID
    - **BirthDate**: 고객 생일
    - **OrderDate**: 주문 날짜
    - **Country**: 주문한 고객의 국가
    - **DiscountRate**: 할인율

2. **매출 분석**

        - 매출액 = `Quantity` * `Price`
    2. 월별 총 매출액을 집계하여 시각화하세요.

```python
# Part A
df['TotalRevenue'] = df['Quantity'] * df['Price']

# DataFrame.resample(rule, axis=0, closed=None, label=None, convention='start', kind=None, loffset=None, base=None, on=None, level=None, origin='start_day', offset=None)
    # rule : 리샘플링 할 기준 입니다. 단위로는 Y, M, D, H, T(min), S ... 등을 조합하여 사용할 수 있습니다.
    # on=index를 입력하여 index라는이름을 가진 열을 기준으로 리샘플링을 진행할 수 있습니다. >>> 아...

# Part B
df_resample = df.resample('ME', on='OrderDate') # on이 핵심이었어..
# print(df_resample.first())

# 미완
# plt.bar(df_resample['OrderDate'], df_resample['TotalRevenue'], zorder=10)
# plt.xlabel('월')
# plt.ylabel('총매출액')
# plt.title('월별 총매출액')
# plt.show()

# Part C (미완)
df_groupby = df.groupby('Category')
category_TR = df_groupby['TotalRevenue'].sum('TotalRevenue')
category_TR.sort_values

```
