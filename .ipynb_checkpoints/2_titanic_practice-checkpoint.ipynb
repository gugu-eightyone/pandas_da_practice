##### [실습] titanic 승객 데이터 활용

**컬럼 설명** 
- PassengerId - 승객 고유번호
- Survived - 생존유무(target 값) (0 = 사망, 1 = 생존)
- Pclass - 티켓 클래스 (1 = 1st, 2 = 2nd, 3 = 3rd)
- Name - 탑승객 성명
- Sex - 성별 (male: 남성, female: 여성)
- Age - 나이(세)
- SibSp - 함께 탑승한 형제자매, 배우자 수 총합
- Parch - 함께 탑승한 부모, 자녀 수 총합
- Ticket - 티켓 넘버
- Fare - 탑승 요금
- Cabin - 객실 넘버
- Embarked - 탑승 항구 (C: Cherbourg, Q: Queenstown, S: Southampton)

```python
# titanic.csv 파일 로드
# titanic 데이터셋의 메타 데이터 확인

import numpy as np
import pandas as pd

df = pd.read_csv('./data/titanic.csv')

print(df.head()) # 어떻게 생겼나 보기
print("============================================")
print(df.info()) # 기본 정보
print("============================================")
print(df.describe())
    # cabin 컬럼에 null값이 엄청 많고
    # 나이 컬럼에 null값이 좀 있고
    # Embarked 컬럼에도 null이 존재함
print("============================================")
df[df.duplicated()] # 중복값 확인
```

- 기초문제

```python
# 1. 전체 승객 중 생존자의 수와 사망자의 수를 구하여 출력

df['Survived'].value_counts()
```

```python
# 2. 승객 나이의 평균 출력

print(df['Age'].mean())
```

```python
# 3. 1등급 승객 중 운임이 가장 높은 승객의 이름 출력

# 초안
# highestfare = df[df['Pclass'] == 1]['Fare'].max()

highestfare = df.loc[df['Pclass'] == 1, 'Fare'].max()
df[ (df['Pclass'] == 1) & (df['Fare'] == highestfare)]
```

```python
# 4. 여성 승객의 수를 구하여 출력

(df['Sex'] == 'female').sum()
```

```python
# 5. 운임이 50 이상인 승객의 이름과 운임 출력

# 초안
# df[df['Fare']>=50][['Name','Fare']]

# .loc[행 조건, 열 조건]
df.loc[df['Fare']>=50 ,['Name','Fare']]
```

```python
# 6. 나이가 18세 미만인 승객들의 생존율을 계산하여 출력

minor_total = len( df[ df['Age'] < 18 ] )
minor_survived = df.loc[(df['Age'] < 18), 'Survived'].sum()
print( minor_survived / minor_total )
```

```python
# 7. 2등급 객실에 탑승한 승객의 평균 운임을 계산하여 출력

df.loc[ df['Pclass'] == 2 , 'Fare' ].mean()
```

```python
# 8. 승객 중 나이가 가장 많은 사람의 나이와 이름을 출력

df.sort_values('Age', ascending = False)

# DataFrame.idxmax(axis=0, skipna=True) >>> 해당 axis 방향으로 최대인 행(열)의 인덱스를 반환
index1 = df['Age'].idxmax()
df.loc[index1, ['Age','Name']]
```

```python
# 9. S 항구에서 탑승한 승객의 수를 출력

# num_C = (df['Embarked'] == "C").sum()
num_S = (df['Embarked'] == "S").sum()
# num_Q = (df['Embarked'] == "Q").sum()
# total = num_C + num_S + num_Q

# print(total)
print(num_S)
```

```python
# 10. 3등급 객실에 탑승한 남성 승객 중 생존자의 수를 출력

condition1 = df['Pclass'] == 3 # 3등급 객실이며
condition2 = df['Sex'] == 'male' # 남성이며
condition3 = df['Survived'] == 1 # 생존자인가

len( df[condition1 & condition2 & condition3] )
```

- 고급문제

```python
# 1. 생존자(Survived)와 그렇지 않은 사람들의 `Pclass`(객실 등급)별 평균 나이를 각각 출력

df_2grouped = df.groupby(['Pclass','Survived'])
df_2grouped['Age'].mean()
```

```python
# 2. `SibSp`(형제/배우자 수)가 3 이상인 승객들 중에서, `Fare`(운임) 상위 5명의 이름과 나이를 출력

df_sibsp3 = df[df['SibSp'] >= 3]
df_sibsp3.sort_values('Fare', ascending=False).head()
```

```python
# 3. 승객 중에서 `Age` 값이 결측치(NaN)인 사람들의 `Embarked`(탑승 항구)별 분포를 출력

df_sub = df[df['Age'].isna()]
# print(df_sub)

df_embarked = df_sub.groupby('Embarked')

df_embarked.size()

# for name,group in df_embarked:
#     print(name)     
#     display(group)
```

```python
# 4. 나이가 16세 미만이고, 혼자 탑승한(`SibSp == 0` and `Parch == 0`) 승객의 이름과 성별, 생존 여부를 출력

condition1 = df['Age'] < 16
condition2 = df['SibSp'] == 0
condition3 = df['Parch'] == 0

df.loc[ condition1 & condition2 & condition3, ["Name","Sex","Survived"] ]
```

```python
# 5. 1등급(Pclass == 1) 객실에 탑승한 승객들 중에서, 운임(Fare)이 평균보다 높은 사람들의 이름과 나이, 운임을 출력

condition1 = df['Pclass'] == 1

mean_val = df.loc[condition1, 'Fare'].mean()
print(mean_val)

condition2 = df.loc[condition1, 'Fare'] > mean_val

df.loc[condition1 & condition2, ['Name','Age','Fare']]
```

```python
# 6. 나이(Age), 객실 등급(Pclass), 운임(Fare) 값에 결측치가 없는 승객들만 대상으로, 각 등급별 평균 운임과 중앙값 운임을 출력

# DataFrame.dropna(axis=0, how='any', thresh=None, subset=None, inplace=False)
    # axis : {0: index / 1: columns} 결측치 제거를 진행 할 레이블
    # how : 'any' : 존재하면 제거 / 'all' : 모두 결측치면 제거
    # tresh : 결측값이 아닌 값이 몇 개 미만일 경우에만 적용시키는 인수
    # subset : dropna메서드를 수행할 레이블을 지정 >>> 특정 열(혹은 행)에서만 결측치를 탐색하도록 할 수 있음 goat..
    # inplace : 원본을 변경할지의 여부

df_dropna = df.dropna(subset=['Age','Pclass','Fare'])
print(df_dropna.info())
print("=============")

df_groupby = df_dropna.groupby('Pclass')

df_groupby['Fare'].mean(), df_groupby['Fare'].median()
```

```python
# 7. 승객들의 `Embarked` 별 평균 나이를 계산하고, 가장 많은 승객이 탑승한 항구를 기준으로 생존율을 계산하여 출력

# 항구별 평균 나이
df_groupby = df.groupby('Embarked')
print(df_groupby['Age'].mean())

# 승객 많은 거 누굴까
# len(df_groupby.get_group('C')), len(df_groupby.get_group('Q')), len(df_groupby.get_group('S'))
    # >>> size!
print(df_groupby.size())

# 생존율 계산
df_S = df_groupby.get_group('S')
print(df_S['Survived'].sum() / len(df_S))
```

```python
# 8. 승객 중에서 여성(`Sex == 'female'`)이면서 30세 이상인 사람들의 생존율을 계산하여 출력

condition1 = df['Sex'] == 'female'
condition2 = df['Age'] >= 30

df_survived = df.loc[condition1 & condition2, 'Survived']
print( df_survived.sum() / len(df_survived) )
```

```python
# 9. 동반한 형제 또는 배우자 수(`SibSp`)와 부모 또는 자녀 수(`Parch`)의 합이 3 이상인 승객들만을 대상으로, 그들의 생존율을 계산하여 출력

df1 = df
df1['altogether'] = df1['SibSp'] + df1['Parch']
# print(df1)

condition = df1['altogether'] >= 3

print( df1.loc[condition,'Survived'].sum() / len(df1.loc[condition]) )
```
